```python
path = 'C:/Users/Wang/datasets/movielens/'
data = pd.read_csv(path+'ratings.csv', sep=',')
```

```python
def get_level(x):
    if 0 <= x and x < 1:
        return 'low'
    elif 1<=x and x <4:
        return 'middle'
    elif x >= 4:
        return 'high'
    else:
        return np.nan
```

```python
data['level'] = data['rating'].map(get_level)
data.head()
```

```python
grouped = data.groupby(by='level')
```

```python
grouped['rating'].value_counts()
```

```python
grouped['rating'].mean()
```

```python
grouped['rating'].std()
```

```python
grouped['rating'].skew()
```

```python
for key, pieace in grouped:
    print(key, len(pieace))
```
