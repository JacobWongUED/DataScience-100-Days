## 1.1 usa.gov

**查看数据格式（json）**

```python
path = "C:/Users/Wang/DataScience-100-Days/datasets/bitly_usagov/example.txt"
open(path).readline()
```

**转换数据格式（json->dict）**

```python
import json
records = [json.loads(line) for line in open(path)] 
records[0]
```

```python
records[0]["tz"]
```

**对时区进行计数**

```python
time_zones = [rec["tz"] for rec in records if "tz" in rec]
time_zones[:10]
```

```python
# 利用pandas
import pandas as pd
import numpy as np

frame = pd.DataFrame(records)
tz_counts = frame["tz"].value_counts()
tz_counts.head()
```

**可视化**

```python
%matplotlib inline
tz_counts[:10].plot.barh(figsize=(10, 10))
```

**获取agent**

```python
frame["a"].head()
```

```python
results = pd.Series([x.split()[0] for x in frame["a"].dropna()])
results.head()
```

**获取操作系统**

```python
filter = np.logical_and(frame["a"].notnull(), frame["tz"].notnull())
clean_frame = frame[filter]
```

```python
operation_system = np.where(clean_frame["a"].str.contains("Windows"), "Windows", "Not Windows")
operation_system[:5]
```

**利用时区和操作系统分组**

```python
by_tz_os = clean_frame.groupby(['tz', operation_system])
```

```python
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

**选取最常出现的时区**

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot.barh(stacked=True, figsize=(10, 10))
```

## 2.6 索引

**布尔型索引**

```python
names = np.array(["Jacob", "Bob", "Max", "Marx", "Bob"])
data = np.random.randn(5, 5)
data
```

```python
data[:3, 2:]
```

```python
data[4]
```

```python
data[2, :2]
```

```python
bob_filter = (names == "Bob")
bob_filter
```

```python
data[bob_filter]
```

```python
data[bob_filter, 2:]
```

```python
other_filter = (names != "Bob")
data[other_filter]
```

```python
data[np.logical_not(bob_filter)]
```

**logical_ot(A == B) 等价于 A != B**

```python
lower_than_zero = data < 0
data[lower_than_zero]
```

```python
bob_or_jacob = (names == "Bob") | (names == "Jacob")
data[bob_or_jacob]
```

**花式索引**

```python
size = (5, 5)
arr = np.empty(size)
for i in range(5):
    arr[i] = i
arr
```

```python
arr[[0, 2, 4]]
```

```python
arr[[4, 2, 0]]
```

```python
arr[[-1, -3]]
```

```python
arr = np.arange(36).reshape(6, 6)
arr
```

```python
arr[[1, 5, 2], [0, 3, 2]]
```

**np.ix_能将两个一维数组转换成选区方形区域的索引**

```python
arr[[1, 5, 2]] [:, [0, 3, 2]]
```

```python
arr[np.ix_([1, 5, 2], [0, 3, 2])]
```

**不同于切片， 花式索引将结果保存在新的数组中**
