# 2.1 数据探查

本数据包含北京、成都、上海、沈阳、广州五城市2010/1/1-2015/12/31的PM2.5以及气象数据，各列含义如下：
> No: 序号  
> year: 年份  
> month: 月份  
> season: 季节  
> PM: PM2.5浓度  
> DEWP(露点): 水蒸气与水达到平衡的温度，用来计算引擎结冰以及出现雾的可能性  
> TEMP: 摄氏温度  
> HUMI: 湿度（%）  
> PRES: 气压（hPa）  
> cbwd: 组合风向  
> lws: 累计风速（m/s）  
> precipitation: 降水量（mm/h）  
> lprec: 累计降水量（mm）
